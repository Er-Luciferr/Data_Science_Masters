<h3>Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.graph_objects as go
import matplotlib.pyplot as plt
data1 = sns.load_dataset('titanic')
data1.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = data1.age , y = data1.fare , mode = 'markers' ))
```

<h3>Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
tip = sns.load_dataset('tips')
import plotly.express as px
fig = px.box(x = tip.total_bill , y = tip.time)
fig.show()
```

<h3>Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
fig = px.histogram( tip , x='sex' , y= 'total_bill' , color='day', pattern_shape='smoker'  )
fig.show()
```

<h3>Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
<br>Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
data4 =sns.load_dataset('iris')
data4.head()
```

```python
fig = px.scatter_matrix( data4 , color='species')
plt.figure(figsize=(12,12))
fig.show()
```

<h3>Q5. What is Distplot? Using Plotly express, plot a distplot.

```python
import plotly.figure_factory as ff
import plotly.express as px

df = px.data.tips()
fig = ff.create_distplot([df['total_bill']], ['total_bill'])
fig.show()
```
