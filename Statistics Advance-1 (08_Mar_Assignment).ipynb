<h3>Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
from math import sqrt, pi, exp

def normal_pdf(x, mu, sigma):
    return (1 / (sigma * sqrt(2 * pi))) * exp(-((x - mu) ** 2) / (2 * sigma ** 2))
```

<h3>Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

1. There are only two possible outcomes for each trial: success or failure.
<br>2. The trials are independent, meaning the outcome of one trial does not affect the outcome of another trial.
<br>3. The probability of success is the same for each trial.
<br>4. The mean of the distribution is equal to the product of the number of trials and the probability of success (μ = np).
<br>5. The variance of the distribution is equal to the product of the number of trials, the probability of success, and the probability of failure (σ² = np(1-p)).
<br>6. The standard deviation of the distribution is equal to the square root of the variance (σ = √np(1-p)) .

Two examples of events where binomial distribution can be applied are:

<br>1. Flipping a coin: If you flip a coin 10 times, you can use the binomial distribution to calculate the probability of getting a certain number of heads.
<br>2. Quality control: If you randomly select 100 items from a production line and test them for defects, you can use the binomial distribution to calculate the probability of finding a certain number of defective items.

<h3>Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(0)

# Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
sample = np.random.binomial(n=1, p=0.4, size=1000)

# Plot a histogram of the sample
plt.hist(sample, bins=[-0.5, 0.5, 1.5], edgecolor='black')
plt.xticks([0, 1])
plt.title('Histogram of Binomial Sample')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

<h3>Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
from math import exp, factorial

def poisson_cdf(k, lam):
    prob = 0
    for i in range(k + 1):
        prob += (lam ** i) * exp(-lam) / factorial(i)
    return prob
```

<h3>Q7. How Binomial distribution different from Poisson distribution?

<br> i) Binomial distribution studies the probability of repeated number of trials with two possible outcomes (success or failure) and two parameters (n and p).
<br> ii) Poisson distribution gives the count of independent events that occur randomly within a given period with unlimited number of possible outcomes and one parameter (m).
