<h3>Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
<br>import numpy as np
<br>np.random.seed(3)
<br>x = 3 + np.random.normal(0, 2, 50)
<br>y = 3 + np.random.normal(0, 2, len(x))
<br>Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y , c='#86813F' , alpha=0.6)
plt.xlabel('This is X-axis')
plt.ylabel('This is Y-axis')
plt.title('Scatter-Plot')
plt.grid()
```

<h3>Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
<br>Use the following data:
<br>import numpy as np
<br>For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
<br>For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
<br>For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
<br>For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,(1))
plt.plot(x1,y1)
plt.title('Line Plot-1')

plt.subplot(2,2,(2))
plt.plot(x2,y2)
plt.title('Line Plot-2')

plt.subplot(2,2,(3))
plt.plot(x3,y3)
plt.title('Line Plot-3')

plt.subplot(2,2,(4))
plt.plot(x4,y4)
plt.title('Line Plot-4')

plt.suptitle('Four Line Plots')
plt.show()
```
